# Projeto de ETL, com ai generativa. Para mensagens

Projeto feito com o google AI


## **E**xtract


```python
import getpass

api_key = getpass.getpass("API KEY: ")
base_url = "https://jsonplaceholder.typicode.com/users?_limit=3"
```

```python
import requests
import json

def get_user():
  response = requests.get(base_url)
  return response.json() if response.status_code == 200 else None

users = get_user()
print(users)
```

## **T**ransform

Usando a lib do google generativeai

```python
!pip install google-generativeai
```

```python
import google.generativeai as genai
import json

genai.configure(api_key=api_key)

model = genai.GenerativeModel("gemini-2.5-flash-lite")

def generate_ai(user):
    print(f"Gerando para: {user['name']}")

    prompt = (
        f"Você é um especialista em marketing bancário. Crie uma mensagem de até 100 caracteres "
        f"para {user['name']} sobre a importância de guardar e investir dinheiro. "
        f"A resposta deve ser descontraída e pronta para WhatsApp. Envie apenas a frase."
    )

    response = model.generate_content(prompt)

    text = response.text.strip() if response.text else ""

    return {
        "text": text,
        "name": user['name'],
        "email": user['email'],
        "link": "https://google.com"
    }

leads = []
for user in users:
    lead = generate_ai(user)
    leads.append(lead)

print(json.dumps(leads, indent=2, ensure_ascii=False))
```
